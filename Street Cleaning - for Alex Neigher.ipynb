```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Street Cleaner Arrival Times - Sheet1.csv')
df.head()
```

```python
plt.hist(df['arrival time'])
plt.show()
```

```python
sns.swarmplot(x='DOW', y='arrival time', data=df)
plt.title('Street cleaning times by day-of-week')
plt.xlabel('Day of week (1=Sunday)')
```

```python
sns.distplot(df[df['DOW'] == 2]['arrival time'], hist=False, label='Monday')
sns.distplot(df[df['DOW'] == 5]['arrival time'], hist=False, label='Thursday')
plt.title('Distribution of street cleaning times by day')
```

```python
sns.jointplot(data=df, x='DOW', y='arrival time', kind='hex')
plt.title('Over engineered graph')
```
